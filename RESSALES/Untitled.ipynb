```python
import pandas as pd
```

```python
#skip rows and seperate all tables
#save each table to individual csv file
f = open('RESSALES-mf.csv').read().split("\n\n\n")
f1 = f[0]
f2 = f[1]
f3 = f[2]
f4 = f[3]
f5 = f[4]

file1 = open('categories.csv','w')
file2 = open('data_types.csv','w')
file3 = open('error_types.csv','w')
file4 = open('geo_levels.csv','w')
file5 = open('time_periods.csv','w')

file1.write(f1)
file2.write(f2)
file3.write(f3)
file4.write(f4)
file5.write(f5)

file1.close()
file2.close()
file3.close()
file4.close()
file5.close()

df_main = pd.read_csv("RESSALES-mf.csv",
                  skiprows = 766)
df_main.to_csv("data_main.csv")
```

```python
#read in all csv files, remove headers
df_categories=pd.read_csv("categories.csv",skiprows=1)
df_data_types=pd.read_csv("data_types.csv",skiprows=1)
df_error_types=pd.read_csv("error_types.csv",skiprows=1)
df_geo_levels=pd.read_csv("geo_levels.csv",skiprows=1)
df_time_periods=pd.read_csv("time_periods.csv",skiprows=1)
df_main=pd.read_csv("data_main.csv",skiprows=1)
```

```python
#merge time and main tables
df = pd.merge(df_time_periods, df_main, on=['per_idx'])
df.drop('per_idx', axis=1, inplace=True)
```

```python
df['per_name'] = pd.to_datetime(df['per_name'])
```
